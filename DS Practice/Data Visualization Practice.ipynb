# Canada Immigration Dataset

```python
df_can = pd.read_csv(r"C:\Users\jj021\Desktop\DataScience\Canada.csv.csv", encoding="Latin-1")
df_can.head()
```

```python
df_can.drop(df_can.columns[43:], axis=1, inplace=True)
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.rename(columns={"OdName":"Country", "AreaName":"Continent", "RegName":"Region"}, inplace=True)
df_can.head()
```

```python
df_can.drop(df_can.index[196:], axis=0, inplace=True)
df_can.shape
```

```python
years=list(df_can.columns[4:])
df_can[years] = df_can[years].astype("int")
df_can["Total"]=df_can.sum(axis=1)
df_can.head()
```

# Transforming Canada Dataset For Visualization

```python
years=list(map(str,range(1980,2014)))
df_can_new=df_can.melt(id_vars=["Country", "Continent", "Region", "DevName"], value_vars=df_can[years], var_name="Year",
                       value_name="Immigrants")
df_can_new["Year"] = df_can_new["Year"].astype("int")
df_can_new.head()
```

# Canada Immigration Dataset Visualizations

### China, India, UK Line Plot

```python
fig = px.line(df_can_new.loc[df_can_new["Country"].isin(["India", "China", "United Kingdom of Great Britain and Northern Ireland"])], 
              x="Year", y="Immigrants", color="Country", title="Immigration Trend From The Countries With The Most Immigrants to Canada")
fig.show()
```

### Immigrants from Eastern Asia Bar Plot

```python
fig = px.histogram(df_can_new[df_can_new["Region"]=="Eastern Asia"], x="Immigrants", y="Country", color="Country",
                   title="Immigrants to Canada From Eastern Asian Countries Comparison", orientation="h")
fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("sum of", "Total")))
fig.update_layout(xaxis_title="Total Immigrants")
fig.show()
```

### Immigrants from Western Europe Bar Plot

```python
fig = px.histogram(df_can_new[df_can_new["Region"]=="Western Europe"], x="Immigrants", y="Country", color="Country",
                   title="Immigrants to Canada from Western European Countries Comparison", orientation="h")
fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("sum of", "Total")))
fig.update_layout(xaxis_title="Total Immigrants")
fig.show()
```

### Total Immigrants By Year Bar Plot

```python
fig = px.histogram(df_can_new, x=df_can_new["Year"].astype("str"), y="Immigrants", title="Total Immigrants to Canada By Year",
                  labels={"x":"Year"})
fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("sum of", "Total")))
fig.update_layout(xaxis_title="Years", yaxis_title="Total Immigrants")
fig.show()
```

### Immigrants from Egypt and Mexico Year by Year Comparison Bar Plot

```python
df_can_new["Year"] = df_can_new["Year"].astype("str")
fig = px.histogram(df_can_new[df_can_new["Country"].isin(["Egypt", "Mexico"])], x="Year", y="Immigrants", color="Country", barmode="group",
                   title="Canadian Immigrants from Egypt vs. Mexico Year by Year Comparison", color_discrete_sequence=["#ff8c00","#008000"])
fig.for_each_trace(lambda t: t.update(hovertemplate=t.hovertemplate.replace("sum of", "")))
fig.update_layout(yaxis_title="Immigrants")
fig.show()
```
